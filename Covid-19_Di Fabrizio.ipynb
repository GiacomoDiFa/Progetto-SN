## Il modello SEIR in breve 
In una popolazione esistono più gruppi di individui, caratterizzati da quattro numeri

* Suscettibili $S$: numero di individui che possono essere infettati 
* Esposti $E$: numero di individui che sono stati infettati ma non sono ancora infettivi 
* Infettivo $I$: numero di individui che sono infettivi 
* Guariti $R$: numero di individui che sono stati contagiati e sono guariti, acquisendo (speriamo) anche un'immunità.

Tutte queste grandezze sono funzione del tempo, che di seguito viene misurato in giorno. Chiameremo $N(t) = S(t) + E(t) + I(t) + R(t)$ il numero totale di individui nella popolazione in funzione del tempo $t$; nel modello questo numero non è costante e il suo valore iniziale è $N(0) = N_0$..

Inoltre, introduciamo diversi parametri
* Il tasso di mortalità naturale $\mu$, ovvero la frazione di popolazione che muore ogni giorno: si assume pari al tasso di natalità, per mantenere stabile la popolazione.
* La forza dell'infezione $\beta$: è legata alla probabilità che una persona suscettibile, venendo a contatto con una persona infetta, venga effettivamente esposta.
* Il ritardo dall'esposizione al diventare infettivo $\epsilon$: è legato al tempo necessario a una persona esposta per diventare infettiva.
* Il tasso di guarigione $\gamma$: l'inverso della durata di una malattia. Assumeremo che una persona infetta sia contagiosa durante tutta la sua malattia. 

Tutte queste ipotesi consentono di modellare l'evoluzione delle diverse componenti della popolazione; per i soggetti suscettibili si ha
\begin{equation}
\frac{d S}{d t} = \mu (N - S) - \beta \frac{I\cdot S}{N}
\end{equation}
dove il primo termine è il saldo tra nascite e morti naturali, il secondo termine è il numero di persone esposte al giorno; ricorda che $N$ non è costante, $N(t) = S(t) + E(t) + I(t) + R(t)$.

Gli individui esposti cambiano nel tempo come 
\begin{equation}
\frac{d E}{d t} = \beta \frac{I\cdot S}{N} - (\mu + \epsilon) E
\end{equation}
in cui il primo è un termine di crescita dovuto a infezioni, il secondo è un termine di esaurimento dovuto a persone che muoiono naturalmente o diventano infettive.

A loro volta gli individui infetti cambiano come 
\begin{equation}
\frac{d I}{d t} = \epsilon E - (\gamma + \mu + \alpha) I
\end{equation}
in cui il primo termine è il numero di esposti che diventano infettivi per unità di tempo, il secondo termine modella la riduzione del numero di individui infetti, perché guariscono (termine $\gamma$) o muoiono per cause naturali o legate al Covid-19.

Finalmente per le persone guarite abbiamo
\begin{equation}
\frac{d R}{d t} = \gamma I - \mu R
\end{equation}

Queste equazioni sono non lineari a causa del termine $\beta \frac{I\cdot S}{N}$, e dobbiamo quindi integrarle numericamente.

Il numero totale di decessi $D_{Covid}(t)$ dovuti a Covid-19 è ovviamente modellato dall'equazione

\begin{equation}
\frac{d D_{Covid}}{d t} = \alpha I
\end{equation}

che può essere confrontato con il numero di decessi dovuti ad altre cause

\begin{equation}
\frac{d D_{Other}}{d t} = \mu N = \mu (S + E + I + R)
\end{equation}

per valutare la gravità dell'epidemia.

## Modelli di parametro

Fisseremo i valori dei parametri, assumendo il tempo misurato in giorni:

* Nascita (o tasso di mortalità) $\mu = \frac{1}{83\times 365}\,\text{day}^{-1}\simeq 3.3\cdot 10^{-5}\,\text{day}^{-1}$
* Forza di infezione $\beta = 0.25\,\text{day}^{-1}$
* Tasso di recupero $\gamma = \frac{1}{14}\,\text{day}^{-1}$ (che corrisponde a un' infezione della durata di due settimane).
* Tasso di ritardo $\epsilon = \frac{1}{3}\,\text{day}^{-1}$ (tre giorni per diventare infettivo)
* Tasso di mortalità $\alpha = 0.0015\,\text{day}^{-1}$ (durante 14 giorni, circa 2\% delle persone infette muore)
* Popolazione iniziale totale $N_0 = 10^7$ (10 milioni di abitanti, come in Lombardia).

Questi valori sono molto incerti e non dovrebbero essere presi sul serio oltre lo scopo dell'esercizio. In particolare, il tasso di mortalità dipende fortemente dalla qualità delle cure ricevute e se il sistema sanitario viene travolto aumenta notevolmente.

Un importante parametro derivato è il numero di riproduzione ${\cal R}_0$ che rappresenta il numero medio di nuove infezioni dovute a un individuo infetto, durante la sua malattia, assumendo che l'intera popolazione sia suscettibile: è quindi dato da
\begin{equation}
{\cal R}_0 = \frac{\beta \epsilon}{\left( \epsilon + \mu\right)\left(\gamma + \alpha + \mu\right)}
\end{equation}
 
Con i parametri ipotizzati, il numero di riproduzione di base è quindi ${\cal R}_0 \simeq 3.4$, simile a quanto osservato al culmine del contagio in Lombardia.

Per controllare il contagio, un governo può ridurre il parametro $\beta$ imponendo un lockdown, come ha fatto l'Italia all'inizio di marzo.

Useremo quindi anche il valore $\beta_{lockdown} = 0.04\,\text{day}^{-1}$, che porta ${\cal R}_0 \simeq 0.55$.

Scrivere un codice che integri numericamente le equazioni per $S, E, I, R, D_{Covid}, D_{Other}$, con uno schema di discretizzazione del 1° ordine nel tempo, ad esempio lo schema di Eulero diretto.

Mantenere la fase di discretizzazione $\Delta t$ distinta dalla fase di campionamento $\Delta t_s$ utilizzata per campionare e salvare i valori per le diverse quantità $N, S, E, I, R, D_{Covid}, D_{Other}$.
```python
from math import ceil, log 
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
#condizioni iniziali
N0 = 10**7
S0 = N0 - 1
E0 = 1
I0 = 0
R0 = 0
D_covid0 = 0
D_other0 = 0

#valori definiti
T = 365
dt = 1
N = int(T/dt)
t = numpy.linspace(0,T,N)
died = numpy.empty(N)            #diedforday senza lockdown
died_lockdown = numpy.empty(N)   #diedforday con lockdown
x = numpy.linspace(0,T,N-1)      #per plot diedforday
```

Utilizzo il metodo di eulero utilizzando questo sistema di equazioni:
\begin{align}
S^{n+1} & = S^n + \Delta t \left(- \mu\,( N^n-S^n) - \beta\, \frac{I^n\cdot S^n}{N^n} \right) \\
E^{n+1} & = E^n + \Delta t \left(\beta \frac{I^n \cdot S^n}{N^n}\ - (\mu + \epsilon)E^n \right) \\
I^{n+1} & = I^n + \Delta t \, [\epsilon E^n - (\gamma + \mu + \alpha) \cdot I^n ] \\
R^{n+1} & = R^n + \Delta t \,(\gamma I^n - \mu R^n). \\
D_{covid}^{n+1} & = D_{covid}^n + \Delta t \alpha I^n \\
D_{other}^{n+1} & = D_{other}^n + \Delta t \mu N^n \\
\end{align}

Scrivo questo sistema di equazioni differenziali come un' equazione vettoriale:
$$u'(t) = f(u)$$

dove

\begin{align}
u & = \begin{pmatrix} S \\ E \\ I \\ R \\ D_{covid} \\ D{other} \end{pmatrix} & f(u) & = \begin{pmatrix} \mu (N - S) - \beta \frac{I\cdot S}{N} \\ \beta \frac{I\cdot S}{N} - (\mu + \epsilon) E \\ \epsilon E - (\gamma + \mu + \alpha) I \\ \gamma I - \mu R \\ \alpha I \\ \mu N  \end{pmatrix}.
\end{align}

```python
def f(u):
    """Restituisce il membro destro del sistema di equazioni SEIRD_covidD_other 
    
    Parametri
    ----------
    u : array di float
        array contenente la soluzione al tempo n.
        
    Restituisce
    -------
    dudt : array di float
        array contenente il sistema di equazioni dato u.
    """
    S = u[0]
    E = u[1]
    I = u[2]
    R = u[3]
    D_covid = u[4]
    D_other = u[5]
    
    return numpy.array([mu*(S+E+I+R-S)-beta*I*S/(S+E+I+R),
                       beta*I*S/(S+E+I+R) - (mu + epsilon)*E,
                       epsilon*E - ((gamma+mu+alpha)*I),
                       gamma*I - mu*R,
                       alpha*I,
                       mu*(S+E+I+R)])
```

```python
def euler_step(u,f,dt):
    """Restituisce la soluzione al passo temporale successivo usando il Metodo di Eulero.
    
    Parametri
    ----------
    u : array di float
        soluzione al passo temporale precedente.
    f : funzione
        funzione per calcolare il membro destro del sistema di equazioni.
    dt : float
        time-increment.
    
    Restituisce
    -------
    u_n_più_1 : array di float
        soluzione approssimata al passo temporale successivo.
    """
    return u + dt * f(u)
```

```python
u = numpy.empty((N,6))
u[0] = numpy.array([S0,E0,I0,R0,D_covid0,D_other0])

for n in range(N - 1):
    u[n+1] = euler_step(u[n], f, dt)
    #print("Numero di morti raggiunto:"+str(u[n+1,4]))#+u[n+1,5])) debug
    #print(str(n)+"-->"+str(died[n])) debug
died = u[1:,4] - u[:-1,4]
```

```python
S = u[:,0]
E = u[:,1]
I = u[:,2]
R = u[:,3]
D_covid = u[:,4]
D_other = u[:,5]
```

```python
pyplot.figure(figsize=(10,10))
pyplot.grid()
pyplot.title('SEIR model %.2f' % T, fontsize=18)
pyplot.plot(t,S, label = "S(t)");
pyplot.plot(t,E, label = "E(t)");
pyplot.plot(t,I, label = "I(t)");
pyplot.plot(t,R, label = "R(t)");
pyplot.plot(t,D_covid, label = "D_covid(t)");
pyplot.plot(t,D_other, label = "D_other(t)");
pyplot.legend()
pyplot.xlabel("Tempo")
pyplot.ylabel("Proporzioni")
```

A tal fine, definire un errore come la differenza nella norma $L_1$ tra i risultati ottenuti con diversi passaggi di integrazione $\Delta t$, e i risultati ottenuti con il passo di integrazione minore, come abbiamo fatto in alcune lezioni.

Verificare che l'errore sia $O(\Delta t)$, cioè decresce linearmente con $\Delta t$ decrescente.

```python
#array contenente i vari valori di dt
dt_values = numpy.array([0.01,0.02,0.05,0.1,0.2,0.5,1.])

#array contenente i vari vettori calcolati
u_values = numpy.empty_like(dt_values, dtype=numpy.ndarray)

for i, dt in enumerate(dt_values):
    
    N = int(T/dt)

    ### discretizzo il tempo in base a ogni dt ###
    t = numpy.linspace(0.0, T, N)
    
    # inizializzo l'array contenente la soluzione per ogni passo temporale
    u = numpy.empty((N, 6))
    u[0] = numpy.array([S0,E0,I0,R0,D_covid0,D_other0])

    # time loop
    for n in range(N-1):
       
        u[n+1] = euler_step(u[n], f, dt)   ### eseguo il metodo di Eulero ###
    
    # salvo il valore di u relativo a una griglia
    u_values[i] = u
```

Ora che abbiamo tutti i vettori opportuni ricordiamo cosa è la norma L-1 per poterla implementare più facilmente
\begin{equation}
E = \Delta t \sum_{n=0}^N \left|z(t_n) - z_n\right| = \frac{T}{N}\sum_{n=0}^N \left|z(t_n) - z_n\right|
\end{equation}

```python
def get_diffgrid(u_current, u_fine, dt):
    """Restituisce la differenza tra una griglia e la griglia piu fine usando la norma L-1.
    
    Parametri
    ----------
    u_current : array di float
        soluzione della griglia corrente.
    u_finest : array di float
        soluzione della griglia fine.
    dt : float
        time-increment della griglia corrente.
    
    Restituisce
    -------
    diffgrid : float
        differenza calcolata nella norma L-1.
    """
    
    N_current = len(u_current[:,0])
    N_fine = len(u_fine[:,0])
   
    grid_size_ratio = int(ceil(N_fine/N_current))
    
    diffgrid = dt * numpy.sum( numpy.abs(\
            u_current[:,2]- u_fine[::grid_size_ratio,2])) 
    
    return diffgrid
```

```python
# calcolo la differenza tra una soluzione a griglia e la piu fina
diffgrid = numpy.empty_like(dt_values)

for i, dt in enumerate(dt_values):
    
    ### eseguo la funzione get_diffgrid() ###
    diffgrid[i] = get_diffgrid(u_values[-1], u_values[i], dt)
```

```python
# log-log plot della differenza tra le griglie
pyplot.figure(figsize=(6,6))
pyplot.grid(True)
pyplot.xlabel('$\Delta t$', fontsize=18)
pyplot.ylabel('$L_1$-norm of the grid differences', fontsize=18)
pyplot.axis('equal')
pyplot.loglog(dt_values[:-1], diffgrid[:-1], color='k', ls='-', lw=2, marker='o');
```

Calcoliamo ora l'ordine di convergenza, ovvero la velocità con cui la soluzione numerica si avvicina a quella esatta man mano che la griglia viene raffinata.
Considerando che non stiamo confrontando con una soluzione esatta, usiamo 3 risoluzioni della griglia che vengono raffinate a un rapporto costante $r$ per trovare *l'ordine di covergenza osservato* ($p$), che è dato da:

\begin{equation}
p = \frac{\log \left(\frac{f_3-f_2}{f_2-f_1} \right) }{\log(r)}
\end{equation}

dove $f_1$ è la soluzione griglia più fine e $f_3$ la più grossolana. 

```python
r = 2
h = 0.001

dt_values2 = numpy.array([h, r*h, r**2*h])

u_values2 = numpy.empty_like(dt_values2, dtype=numpy.ndarray)

diffgrid2 = numpy.empty(2)

for i, dt in enumerate(dt_values2):
    
    N = int(T/dt)   # numero di passi temporali
    
    ### discretizzo il tempo t ###
    t = numpy.linspace(0.0, T, N)
    
    # inizializzo l'array contenente la soluzione per ogni passo temporale
    u = numpy.empty((N, 6))
    u[0] = numpy.array([S0,E0,I0,R0,D_covid0,D_other0])

    # loop temporale
    for n in range(N-1):

        u[n+1] = euler_step(u[n], f, dt)         ### eseguo euler_step() ###
    
    # salvo il valore di u relativo a una girglia
    u_values2[i] = u
    

#calcolo f2 - f1
diffgrid2[0] = get_diffgrid(u_values2[1], u_values2[0], dt_values2[1])

#calcolo f3 - f2
diffgrid2[1] = get_diffgrid(u_values2[2], u_values2[1], dt_values2[2])

# calcolo l'ordine di convergenza
p = (log(diffgrid2[1]) - log(diffgrid2[0])) / log(r)

print('L ordine di convergenza è p = {:.3f}'.format(p));
```

Verificare che la convergenza sia effettivamente del secondo ordine al diminuire di $\Delta t$, ovvero che l'errore sia $O(\Delta t^2)$, ovvero decresce quadraticamente al diminuire di $\Delta t$.

Ho deciso di utulizzare il metodo di Runge Kutta del secondo ordine o metodo di Eulero modificato.
\begin{align}
u_{n+1/2}   & = u_n + \frac{\Delta t}{2} f(u_n) \\
u_{n+1} & = u_n + \Delta t \,\, f(u_{n+1/2}).
\end{align}

```python
def rk2_step(u_rk2, f_rk2, dt):
    """Restituisce la soluzione al passo temporale successivo usando Runge-Kutta.
    
    Parametri
    ----------
    u_rk2 : array di float
        soluzione al passo temporale precedente.
    f_rk2 : funzione
        funzione per calcolare il membro destro del sistema di equazioni.
    dt : float
        time-increment.
    
    Restituisce
    -------
    u_rk2_n_più_1 : array di float
        soluzione al passo temporale successivo.
    """
    u_star_rk2 = u_rk2 + 0.5*dt*f_rk2(u_rk2)
    return u_rk2 + dt*f_rk2(u_star_rk2)
```
